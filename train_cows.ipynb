# Cow Pose Estimation Training

This notebook trains a YOLOv11 pose model to detect cow keypoints based on the converted Label Studio annotations.

```python
!pip install ultralytics
```

```python
from ultralytics import YOLO
import os

# 1. Load a pretrained YOLOv11 pose model
model = YOLO('yolo11x-pose.pt')

# 2. Train the model using our custom dataset
# We use the coco8-pose.yaml which we updated with 9 keypoints
results = model.train(
    data='coco8-pose.yaml', 
    epochs=4,
    imgsz=640, 
    device='0' # Change to 0 if you have a GPU
)
```

## Inference

After training, you can run inference on an image to see the results.

```python
# Load the best model from training (it will be in runs/pose/train/weights/best.pt)
# For now, let's just use the current model object

test_image = 'images/train/9ce2b0c7-20260101_205224_baia20_IPC2.jpg'
results = model(test_image)

# Show the results
for r in results:
    r.show() # This will open a window with the image and keypoints
```
